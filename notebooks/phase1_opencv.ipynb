```python
import cv2 

# load image
img = cv2.imread(r'D:\Projects\anpr-system\data\test_car.jpg')


# Basic info 
print("Shape:", img.shape)
print("Height:", img.shape[0])
print("Width:", img.shape[1])
print("channels:", img.shape[2])
```

```python
# show img 
cv2.imshow("Car Image", img)

# keeps window open
cv2.waitKey(0)

# close window
cv2.destroyAllWindows()
```

```python
# coverting BGR to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print("Original shape:", img.shape)
print("Grayscale shape:", gray.shape)

cv2.imshow("Grayscale", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Resize
resized = cv2.resize(gray, (0,0), fx=2, fy=2)

# blur
blurred = cv2.GaussianBlur(resized, (5,5), 0)

# threshold
_, thresh = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

cv2.imshow("Resized", resized)
cv2.imshow("Blurred", blurred)
cv2.imshow("Threshold", thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# image save 
cv2.imwrite(r'D:\Projects\anpr-system\data\processed_car.jpg', thresh)
print("Saved!")
```
