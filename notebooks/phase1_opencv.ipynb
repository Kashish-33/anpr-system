# Phase 1 — OpenCV Fundamentals

## Concepts covered
- Image load, display, save
- BGR vs RGB vs Grayscale
- Preprocessing pipeline: resize → blur → threshold

## Key learnings
- Grayscale mein channels nahi hote — shape (H, W) hoti hai
- Kernel size odd hona chahiye GaussianBlur mein
- OTSU automatically best threshold value find karta hai

```python
import cv2 

# load image
img = cv2.imread(r'D:\Projects\anpr-system\data\test_car.jpg')


# Basic info 
print("Shape:", img.shape)
print("Height:", img.shape[0])
print("Width:", img.shape[1])
print("channels:", img.shape[2])
```

```python
# show img 
cv2.imshow("Car Image", img)

# keeps window open
cv2.waitKey(0)

# close window
cv2.destroyAllWindows()
```

```python
# coverting BGR to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print("Original shape:", img.shape)
print("Grayscale shape:", gray.shape)

cv2.imshow("Grayscale", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Resize
resized = cv2.resize(gray, (0,0), fx=2, fy=2)

# blur
blurred = cv2.GaussianBlur(resized, (5,5), 0)

# threshold
_, thresh = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

cv2.imshow("Resized", resized)
cv2.imshow("Blurred", blurred)
cv2.imshow("Threshold", thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# image save 
cv2.imwrite(r'D:\Projects\anpr-system\data\processed_car.jpg', thresh)
print("Saved!")
```

```python
import cv2

# 0 = default camera
cap = cv2.VideoCapture(0)

print("Camera opened:", cap.isOpened())
```

```python
while True:
    # take a frame from webcam
    ret, frame = cap.read()

    # ret -- True/False
    # frame -- actual image (numpy array)

    if not ret:
        print(" Frame not captured")
        break

    cv2.imshow("Live Feed", frame)

    # stops on pressing q
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Camera release 
cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    if not ret:
        break

    # preprocessing pipeline on each frame 
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5,5), 0)
    _, thresh = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # displaying all three
    cv2.imshow("Original", frame)
    cv2.imshow("Gray", gray)
    cv2.imshow("Threshold",thresh)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    cap.release()
    cv2.destroyAllWindows()
```
